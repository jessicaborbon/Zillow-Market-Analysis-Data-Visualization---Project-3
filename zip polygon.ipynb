```python
import json
import pandas as pd


with open('AZ.min.json', 'r') as file:
    data = json.load(file)


zip_ids = [
    "85339", "85310", "85254", "85085", "85054", "85053", "85051", "85050", "85048", "85045",
    "85044", "85043", "85042", "85041", "85040", "85037", "85035", "85034", "85033", "85032",
    "85031", "85029", "85028", "85027", "85024", "85023", "85022", "85021", "85020", "85019",
    "85018", "85017", "85016", "85015", "85014", "85013", "85012", "85009", "85008", "85007",
    "85006", "85004", "85003"
]

```

```python

filtered_data = {}
for feature in data['features']:
    zip_id = feature['properties']['ZCTA5CE10']
    if zip_id in zip_ids:
        # Assuming there's only one list of coordinates and removing one layer of nesting
        filtered_data[zip_id] = feature['geometry']['coordinates'][0]

rows = []
for zip_id in zip_ids:
    if zip_id in filtered_data:
        coordinates = filtered_data[zip_id]
        rows.append([zip_id, coordinates])
mapPLY = pd.DataFrame(rows, columns=['ZIP Code', 'Coordinates'])
mapPLY
```

```python

mapPLY.to_csv('mapPLY.csv', index=False)
```
